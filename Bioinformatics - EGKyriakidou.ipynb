```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.impute import SimpleImputer, KNNImputer
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from sklearn.metrics import confusion_matrix, classification_report, roc_curve
from imblearn.over_sampling import SMOTE
from sklearn.feature_selection import RFE

# Load the dataset
df = pd.read_csv("risk_factors_cervical_cancer.csv")

# Display the basic information
print(df.info())

# Display column labels, to check what the symptoms and details are
print(df.columns)

# Display descriptive statistics for numerical features in case we wish to find details on the numerical characteristics
print(df.describe())

# Display first few rows of the dataset. If we wish to set a sepcific number we can add a number in the parenthesis
print(df.head())

# Check for missing values and print the columns with missing values 
nulls = df.isnull().sum()
print(nulls[nulls != 0])

# Check for non-numerical values and print columns with non-numerical values
non_numeric = df.select_dtypes(exclude=['number']).columns
print(non_numeric)

# Print non-numerical values found in each column
for column in non_numeric:
    non_numeric_values = df[column][df[column].apply(lambda x: pd.to_numeric(x, errors='coerce')).isna()]
    if not non_numeric_values.empty:
        print("Non-numerical values in {column}:")
        print(non_numeric_values)

# Replace all questionmarks and missing values with the KNN
df.replace('?', np.nan, inplace=True) 
imputer = KNNImputer(n_neighbors=5) 
df_imputed = pd.DataFrame(imputer.fit_transform(df), columns=df.columns)

for target in ['Hinselmann', 'Schiller', 'Citology', 'Biopsy']:   
    
    # Feature scaling, to make comparison and colelctive use possible and split the dataset
    scaler = StandardScaler()
    x = scaler.fit_transform(df_imputed.drop(columns=['Hinselmann', 'Schiller', 'Citology', 'Biopsy']))  # Scaling features except the target variable
    y = df_imputed[target]

    # Hanlde imbalanced data
    smote = SMOTE(random_state=42)
    x_resampled, y_resampled = smote.fit_resample(x, y)

    # Feature Selection using RFE with RandomForest
    rf_classifier = RandomForestClassifier(random_state=42)
    rfe = RFE(estimator=rf_classifier, n_features_to_select=10)
    rfe.fit(x_resampled, y_resampled)
    selected_features = df_imputed.drop(columns=['Hinselmann', 'Schiller', 'Citology', 'Biopsy']).columns[rfe.support_]
    print("Most Important Features for", target, ":", selected_features)

    # Feature selection
    rf_classifier.fit(x_resampled, y_resampled)
    feature_importance = rf_classifier.feature_importances_

    plt.figure(figsize=(10, 6))
    plt.barh(df_imputed.drop(columns=['Hinselmann', 'Schiller', 'Citology', 'Biopsy']).columns, feature_importance)
    plt.xlabel('Feature Importance')
    plt.ylabel('Features')
    plt.title('Feature Importance Plot')
    plt.show()
    
    # Split data into training and testing sets
    x_train, x_test, y_train, y_test = train_test_split(x_resampled, y_resampled, test_size=0.2, random_state=42)

    # Models to evaluate
    models = [
        ('Logistic Regression', LogisticRegression(max_iter=1000)),
        ('Gradient Boosting', GradientBoostingClassifier(random_state=42)),
        ('AdaBoost', AdaBoostClassifier(random_state=42)),
        ('Support Vector Machine', SVC(probability=True))
    ]

    # Evaluate each model
    train_results = []
    for name, model in models:
        model.fit(x_train, y_train)
        y_train_pred = model.predict(x_train)  # Predict on training data
        train_accuracy = accuracy_score(y_train, y_train_pred)
        train_precision = precision_score(y_train, y_train_pred)
        train_recall = recall_score(y_train, y_train_pred)
        train_f1 = f1_score(y_train, y_train_pred)
        train_roc_auc = roc_auc_score(y_train, model.predict_proba(x_train)[:, 1])
        train_results.append((name, train_accuracy, train_precision, train_recall, train_f1, train_roc_auc))
        print(f"\nTraining Metrics for {name}:")
        print(f"Accuracy: {train_accuracy:.4f}, Precision: {train_precision:.4f}, Recall: {train_recall:.4f}, F1 Score: {train_f1:.4f}, ROC AUC: {train_roc_auc:.4f}")
        
    # Choose the best model based on training performance
    best_model_name = max(train_results, key=lambda x: x[-1])[0]
    best_model = [model for model_name, model in models if model_name == best_model_name][0]

    # Hyperparameter tuning
    param_grid = {
        'n_estimators': [50, 100, 150],
        'max_depth': [None, 10, 20],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4]
    }

    # Save the best model 
    grid_search = GridSearchCV(estimator=best_model, param_grid=param_grid, cv=5, scoring='roc_auc')
    grid_search.fit(x_train, y_train)
    
    # Cross-Validation
    cv_scores = cross_val_score(grid_search.best_estimator_, x_train, y_train, cv=5, scoring='recall')
    mean_cv_roc_auc = np.mean(cv_scores)

    # Model Evaluation Metrics
    print("\nBest Model Parameters:", grid_search.best_params_)
    print("Cross-Validation ROC AUC Scores:", cv_scores)
    print("Mean CV ROC AUC:", np.mean(cv_scores))

    # Model Evaluation Metrics on test set
    y_pred = grid_search.best_estimator_.predict(x_test)

    # Calculate True Positives (TP), True Negatives (TN), False Positives (FP) and False Negatives (FN)
    TP = np.sum((y_pred == 1) & (y_test == 1))
    TN = np.sum((y_pred == 0) & (y_test == 0)) 
    FP = np.sum((y_pred == 1) & (y_test == 0))
    FN = np.sum((y_pred == 0) & (y_test == 1))

    # Print the evaluation metrics
    print("True Positives (TP):", TP)
    print("True Negatives (TN):", TN)
    print("False Positives (FP):", FP)
    print("False Negatives (FN):", FN)

    # Model Evaluation Metrics on test set
    test_accuracy = accuracy_score(y_test, y_pred)
    test_precision = precision_score(y_test, y_pred)
    test_recall = recall_score(y_test, y_pred)
    test_f1 = f1_score(y_test, y_pred)
    test_roc_auc = roc_auc_score(y_test, best_model.predict_proba(x_test)[:, 1])
    test_roc_auc = roc_auc_score(y_test, grid_search.best_estimator_.predict_proba(x_test)[:, 1])

    print("\nModel Evaluation Metrics on Test Set:")
    print(f"Accuracy: {test_accuracy:.4f}, Precision: {test_precision:.4f}, Recall: {test_recall:.4f}, F1 Score: {test_f1:.4f}, ROC AUC: {test_roc_auc:.4f}")

    # Confusion Matrix
    conf_matrix = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(8, 5))
    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.title('Confusion Matrix')
    plt.show()
```
